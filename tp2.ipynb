## 1.1 Contextualização

A tendência industrial atual em relação aos automatismos e às plantas industriais nos leva a sistemas mecatrônicos cada vez mais complexos, trabalhando em um ambiente incerto e evolutivo. É tão necessário desenvolver um módulo de diagnóstico para detectar uma falha (Fault Detection) que pode afetar a operação desses sistemas e localizar suas causas (Fault Isolation). Portanto, um módulo de diagnóstico é necessário para melhorar o desempenho e a produtividade dos sistemas e limitar as consequências de falhas que podem ser catastróficas para os bens e a vida humana.

## 1.2 Motivação

Para atingir os objetivos de melhoria do desempenho e a produtividade dos sistemas, é necessário ter um sistema eficiente e confiável para detectar e localizar as falhas. O módulo de diagnóstico é um caminho importante para alcançar essas metas.

Vamos usar machine learning para auxiliar o desenvolvimento do módulo de diagnóstico.

## 1.3 Importância

O módulo de diagnóstico é uma ferramenta crucial para detectar e localizar falhas nos sistemas mecatrônicos. Ele pode ajudar a reduzir o número de erros, aumentar a eficiência e a produtividade dos sistemas, além de melhorar a qualidade e a confiabilidade dos bens e a vida humana.
